```python
BACH = "35041-pdf.pdf"
ASTRONOMY = "8172-pdf.pdf"
```

```python
import pymupdf4llm

bach = pymupdf4llm.to_markdown(BACH)
astronomy = pymupdf4llm.to_markdown(ASTRONOMY)

# now work with the markdown text, e.g. store as a UTF8-encoded file
import pathlib
pathlib.Path("bach.md").write_bytes(bach.encode())
pathlib.Path("astronomy.md").write_bytes(astronomy.encode())
```
